```python
import pandas as pd                        
from pytrends.request import TrendReq
pytrends = TrendReq()
```

This is creating a table of interest in Taylor Swift by region. 

```python
pytrend.build_payload(kw_list=["Taylor Swift"])
# Interest by Region
df = pytrend.interest_by_region()
df.head(5)
```

This is creating a table of interest in Taylor Swift by region, no longer inclusing regions with no (0) interest

```python
pytrend.build_payload(kw_list=["Taylor Swift"])
# Interest by Region
df = pytrend.interest_by_region()
nozero = (df != 0).any(axis=1)
clean_df = df.loc[nozero]
clean_df.head(5)
```

This is creating a bar chart of regional interest of Taylor Swift, excluding regions with no interest.

```python
clean_df.reset_index().plot(x="geoName", y="Taylor Swift", figsize=(120, 10), kind ="bar")
```

This is creating a table of regions that have an interest over 50

```python
pytrend.build_payload(kw_list=["Taylor Swift"])
# Interest by Region
df = pytrend.interest_by_region()
overfifty = (df > 50).any(axis=1)
large_df = df.loc[overfifty]
large_df.head(10)
```

```python
large_df.reset_index().plot(x="geoName", y="Taylor Swift", figsize=(120, 10), kind ="bar")
```
